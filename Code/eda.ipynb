```python
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import matplotlib.image as mpimg
from visualizations import sample_augmentation
from config import train_dir, test_dir
```

```python
folders = []
image_counts = []

for folder_name in os.listdir(train_dir):
    folder_path = os.path.join(train_dir, folder_name)
    if os.path.isdir(folder_path):
        num_images = len(os.listdir(folder_path))
        folders.append(folder_name)
        image_counts.append(num_images)
group_counts = dict(zip(folders, image_counts))
group_counts_cleaned = group_counts
group_counts_cleaned = {key.title(): value for key, value in group_counts.items()}
# group_counts_cleaned = {name.title(): count for name, count in group_counts.items() if count >= 500}
```

```python
group_counts_cleaned
```

```python
plt.figure(figsize=(10, 6))
plt.bar(group_counts_cleaned.keys(), group_counts_cleaned.values(), color='skyblue')
plt.xlabel('Dog Groups')
plt.ylabel('Number of Breeds')
plt.title('Number of Breeds per Dog Group')
plt.xticks(rotation=45, ha='right')
ax = plt.gca()
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
plt.tight_layout()
plt.show()
```

```python
working_dir = test_dir + "/working/"
folders = []
image_counts = []

for folder_name in os.listdir(working_dir):
    folder_path = os.path.join(working_dir, folder_name)
    if os.path.isdir(folder_path):
        num_images = len(os.listdir(folder_path))
        folders.append(folder_name)
        image_counts.append(num_images)
group_counts = dict(zip(folders, image_counts))
group_counts_cleaned = {key.split('-')[1].title(): value for key, value in group_counts.items()}
# group_counts_cleaned = {name.title(): count for name, count in group_counts.items() if count >= 500}

sorted_keys = sorted(group_counts_cleaned.keys())
sorted_values = [group_counts_cleaned[key] for key in sorted_keys]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(sorted_keys, sorted_values, color='skyblue')
plt.xlabel('Dog Breeds')
plt.ylabel('Number of Images')
plt.title('Number of Images per Dog Breed for Working Group')
plt.xticks(rotation=45, ha='right')
# plt.xticks()
plt.tight_layout()
plt.show()
```

```python


image_dir = "./data/train/sporting/n02099601-golden_retriever/"
image_files = ["n02099601_9301.jpg",
               "n02099601_3327.jpg",
               "n02099601_2980.jpg"]

descriptions = ["Playing with Ball", "Swimming in Pool", "Hugged by Owner"]

fig, axes = plt.subplots(1, len(image_files), figsize=(15, 5))

for i, (image_file, description) in enumerate(zip(image_files, descriptions)):
    img_path = os.path.join(image_dir, image_file)
    img = mpimg.imread(img_path)
    
    axes[i].imshow(img)
    axes[i].axis('off')     
    axes[i].text(0.5, -0.1, description, size=12, ha='center', transform=axes[i].transAxes)
plt.suptitle("Different Poses of the Golden Retriever", fontsize=16)
plt.tight_layout()
plt.show()
```
